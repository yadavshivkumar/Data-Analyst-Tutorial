```python
## if statement in python is used to test a specific condition. If the condition is true, then the block of code inside the if statement will be executed. If the condition is false, then the block of code inside the if statement will be skipped.   

age = int(input("Enter your age: "))
if age >= 18:
    print("You are eligible to vote.")
    
```

```python
## else statement in python is used to execute a block of code when the condition in the if statement is false. The else statement is optional and can be used with an if statement to provide an alternative block of code to execute when the condition is false.

age = int(input("Enter your age: "))    
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
# elif statement in python is used to test multiple conditions. The elif statement is short for "else if" and can be used to test additional conditions after the initial if statement. If the condition in the if statement is false, the program will check the condition in the elif statement. If the condition in the elif statement is true, the block of code inside the elif statement will be executed. If the condition in the elif statement is false, the program will check any additional elif statements or execute the block of code inside the else statement if it exists.    

age = int(input("Enter your age: "))
if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
else:
    print("You are an adult.")  
```

```python
year = int(input("Enter a year: "))
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0: 
            print(year, " is a leap year.")
        else:
            print(year, " is not a leap year.")
    else:
        print(year, "is not a leap year.")
else:
    print(year, "is not a leap year.")
```
