```python
## DataTypes in Python are used to define the type of data that a variable can hold. Python has several built-in data types, including: 


## Integer: Represents whole numbers, e.g., 1, 42, -5.
age = 25  # Integer data type
type_of_age = type(age) 
print(f"The type of age is: {type_of_age}")  # Output: <class 'int'>
```

```python
##floating data type: Represents decimal numbers, e.g., 3.14, -0.001.
pi = 3.14  # Float data type
type_of_pi = type(pi)
print(f"The type of pi is: {type_of_pi}")  # Output: <class 'float'>
```

```python
## string data type: Represents sequences of characters, e.g., "Hello", 'Python'.
name = "Hello"  # String data type
type_of_name = type(name)       
print(f"The type of name is: {type_of_name}")  # Output: <class 'str'>  
```

```python
## boolean data type: Represents truth values, either True or False.
is_student = True  # Boolean data type
type_of_is_student = type(is_student)
print(f"The type of is_student is: {type_of_is_student}")  # Output: <class 'bool'>
```

```python
## common errors:
# 1. TypeError: This error occurs when an operation or function is applied to an object of an inappropriate type. For example, trying to add a string and an integer will raise a TypeError.


result = "Hello" + str(34)  # This will raise a TypeError because you cannot concatenate a string and an integer.     

print(result)  # Output: Hello 34
```
