```python
## Dictionary is a collection of key-value pairs. Each key is unique and maps to a value. Dictionaries are mutable, meaning you can change their content without changing their identity.   

## it is defined using curly braces {} and key-value pairs are separated by commas. The keys and values can be of any data type, but keys must be immutable (like strings, numbers, or tuples). 

empty_dict = {}  # An empty dictionary
print(empty_dict)  # Output: {}
print(type(empty_dict))  # Output: <class 'dict'>
```

```python
empty_dict = dict()  # Another way to create an empty dictionary
empty_dict
```

```python
student = {
    "name": "John Doe",
    "courses": ["Math", "Science", "History"],
    "is_enrolled": True
}   

print(student)  
print(type(student))  # Output: <class 'dict'>
```

```python
print(student["name"])  # Output: John Doe
print(student["age"])   # Output: 20
print(student["courses"])  # Output: ['Math', 'Science', 'History']
print(student["is_enrolled"])  # Output: True
```

```python
## accessing a value in a dictionary using the get() method. The get() method returns the value for the specified key if the key is in the dictionary. If the key is not found, it returns None (or a specified default value).  

print(student.get("name"))  # Output: John Doe
print(student.get("age"))   # Output: 20
print(student.get("courses"))  # Output: ['Math', 'Science', 'History']
print(student.get("is_enrolled"))  # Output: True
print(student.get("height"))  # Output: None
```

```python
# modifying a value in a dictionary using the key. You can assign a new value to an existing key, or add a new key-value pair if the key does not exist.    

student["age"] = 21  # Modifying an existing key
student["height"] = 5.9  # Adding a new key-value pair  
```

```python
keys = student.keys()  # Returns a view object of the dictionary's keys
print(keys)  # Output: dict_keys(['name', 'courses', 'is_enrolled', 'age', 'height'])
values = student.values()  # Returns a view object of the dictionary's values
print(values)  # Output: dict_values(['John Doe', ['Math', 'Science', 'History'], True, 21, 5.9])
items = student.items()  # Returns a view object of the dictionary's key-value pairs
print(items)  # Output: dict_items([('name', 'John Doe'), ('courses', ['Math', 'Science', 'History']), ('is_enrolled', True), ('age', 21), ('height', 5.

```

```python
# shallow copy of a dictionary using the copy() method. This creates a new dictionary with the same key-value pairs as the original, but they are two separate objects in memory. Changes to one will not affect the other. 

student_copy  = student
print(student)
print(student_copy)
```

```python
for key in student:
    print(f"{key}: {student[key]}")  # Output: name: John Doe, courses: ['Math', 'Science', 'History'], is_enrolled: True, age: 21, height: 5.9 
```

```python
for values in student.values():
    print(values)  # Output: John Doe, ['Math', 'Science', 'History'], True, 21, 5.9
```

```python
for key, value in student.items():
    print(f"{key}: {value}")    
```

```python
# nested dictionaries are dictionaries that contain other dictionaries as values. This allows for more complex data structures and can be useful for representing hierarchical data.    

student = {
    "student1": {
        "name": "John Doe",
        "age": 21,
        "courses": ["Math", "Science", "History"]
    },
    "student2": {
        "name": "Jane Smith",
        "age": 22,
        "courses": ["English", "Art", "Biology"]
    }       
}

print(student["student1"]["name"])  # Output: John Doe
print(student["student2"]["courses"])  # Output: ['English', 'Art', 'Biology']
```

```python
for student_id, student_info in student.items():
    print(f"{student_id}: {student_info['name']}, Age: {student_info['age']}, Courses: {', '.join(student_info['courses'])}")   
```

```python
# dictionary comprehension is a concise way to create dictionaries. It consists of an expression pair (key: value) followed by a for statement inside curly braces {}.

squares = {x: x**2 for x in range(5)}  # Creates a dictionary of squares from 0 to 4
print(squares)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}       
```

```python
evens = {x: x for x in range(10) if x % 2 == 0}  # Creates a dictionary of even numbers from 0 to 9
print(evens)  # Output: {0: 0, 2: 2
```

```python
# pactical example

numbers = [1, 2, 3, 4, 5]
frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1
print(frequency)  # Output: {1: 1, 2: 1, 3: 1, 4: 1, 5: 1}  
```

```python
dict1={ "a" :1, "b":2, "c":3}
dict2={ "a" :1, "b":2, "c":3}   
merged_dict = {**dict1, **dict2}  # Merges dict1 and dict2
print(merged_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}
```
